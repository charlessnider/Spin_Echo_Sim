```python
from data_loader import load_data, load_params
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi as Ï€

# load the parameters
params = load_params()
```

# Your output

```python
# load the echo output
recho, iecho, zrecho, ziecho = load_data("output")

# get all the planar echoes
echoes, zechoes = [], []
for idx in range(len(recho)):
    
    # save the echo as a complex value
    echoes.append(np.add(np.array(recho[idx]), 1j * np.array(iecho[idx])))
    zechoes.append(np.add(np.array(zrecho[idx]), 1j * np.array(ziecho[idx])))


# prepare the number of timepoints
timepoints = []
for param in params:
    
    # 3 times the echo time, + 2 points for the two echoes
    nt = 3 * int(2 * Ï€ * 1000000 * param["Ï„"] / param["dt"]) + 2
    t0 = 0
    tf = param["Ï„"] * 1000000 * 3
    timepoints.append(np.linspace(t0, tf, nt))
    
# some formatting choices
plt.rcParams.update({'font.size': 15})
plt.rcParams['figure.figsize'] = [10, 10]
    
# plot the results
fig, ax = plt.subplots(3, 3)
for idx in range(3):
    for jdx in range(3):
        
        kdx = 3 * idx + jdx
        ax[idx, jdx].plot(timepoints[kdx], np.absolute(echoes[kdx]), linewidth=2.0, c="black")
        ax[idx, jdx].plot(timepoints[kdx], np.real(echoes[kdx]), linewidth=2.0, c="red", alpha=0.25)
        ax[idx, jdx].plot(timepoints[kdx], np.imag(echoes[kdx]), linewidth=2.0, c="blue", alpha=0.25)
        
        # set a title with the params
        title = f"Ï‰p = {round(params[kdx]['Ï‰p'], 1)}Î“, Î¾ = {round(params[kdx]['Î¾'] * 2, 1)}"
        ax[idx, jdx].title.set_text(title)
        
fig.text(0.0, 0.5, "Amplitude (a.u.)", va="center", rotation="vertical")
fig.text(0.5, -0.025, "Time (us)", ha="center")
fig.tight_layout(pad=0.5)
plt.show()
```
